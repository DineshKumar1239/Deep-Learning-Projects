```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Assuming combined_features is a Pandas Series
combined_features = df['genres'] + ' ' + df["keywords"] + " " + df["tagline"] + " " + df["cast"] + " " + df["director"]

# Convert the Pandas Series to a list of strings
combined_features_list = combined_features.astype(str).tolist()

# Create a TfidfVectorizer instance
vectorizer = TfidfVectorizer()

# Use fit_transform on the list of raw text documents
feature_vector = vectorizer.fit_transform(combined_features_list)
```

```python
print(feature_vector)
```

```python
similarity=cosine_similarity(feature_vector)
```

```python


import difflib

# Assuming df is your DataFrame
movie_name = input('Enter your favorite movie name: ')

# Remove NaN values from the 'title' column
list_of_all_titles = df['title'].dropna().tolist()

# Find close matches
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

print(find_close_match)


close_match = find_close_match[0]

index_of_the_movie = df[df.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
